```python
import torch
from torch.autograd import Variable
```

```python
x = Variable(torch.ones(2, 2), requires_grad=True)
print(x)
```

```python
y = x + 2
print(y)
```

```python
print(y.creator)
```

```python
z = y * y * 3
out = z.mean()

print(z, out)
```

```python
out.backward()
```

```python
print(x.grad)
```

```python
x = torch.randn(3)
x = Variable(x, requires_grad=True)
y = x * 2
while y.data.norm() < 1000:
    y = y * 2

print(y)
```

```python
gradients = torch.FloatTensor([0.1, 1.0, 0.001])
y.backward(gradients)

print(x.grad)
```
